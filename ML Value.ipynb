```python
import simfin as sf

sf.set_data_dir('data')
sf.set_api_key('70242009-a4b9-4326-bf75-72be25347d70')
#income sheet, cash flow and balance
data1 = sf.load_income(variant='annual', market='us').copy()
data1.reset_index(inplace=True)

data2 = sf.load_cashflow(variant='annual', market='us').copy()
data2.reset_index(inplace=True)

data3 = sf.load_balance(variant='annual', market='us').copy()
data3.reset_index(inplace=True)

hub = sf.StockHub(market='us')
data4 = hub.val_signals().copy()
data4.reset_index(inplace=True)
data5 = hub.val_signals(variant='latest').copy()
data5.reset_index(inplace=True)

```

```python
from simfin.names import *
rows_to_drop = ['SimFinId', 'Currency', 'Fiscal Period', 'Fiscal Year', 'Publish Date', 'Shares (Diluted)']
def drop_rows(df):
    return df.drop(rows_to_drop, axis=1)

df1 = drop_rows(data1)
df2 = drop_rows(data2)
df3 = drop_rows(data3)
df4 = data4[[TICKER, DATE, MARKET_CAP]]
df4 = df4.reset_index()

df4.columns

```

```python
import pandas as pd
super_df = df1.merge(df2, on=[TICKER, REPORT_DATE])
super_df = super_df.merge(df3, on=[TICKER, REPORT_DATE])
# super_df = pd.concat([df1, df2, df3], join='inner', axis=1)
super_df = super_df.merge(df4, left_on=[TICKER, REPORT_DATE], right_on=[TICKER, DATE])
super_df.columns
```

```python
sfc_df = super_df.drop('Shares (Basic)_x', axis=1)
sfc_df.drop(super_df.filter(regex='_y$').columns.tolist(), axis=1, inplace=True)

dummy = pd.get_dummies(sfc_df[REPORT_DATE])
sfc_df = pd.concat([sfc_df, dummy], axis=1)
# del sfc_df[REPORT_DATE]

backtest_data = sfc_df[sfc_df[REPORT_DATE].dt.year == 2018]
ml_data = sfc_df[sfc_df[REPORT_DATE].dt.year != 2018]
del ml_data[TICKER], ml_data[REPORT_DATE] , backtest_data[REPORT_DATE] 
ml_data = ml_data.fillna(0)
ml_data
```

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
model = RandomForestRegressor()
y = ml_data[MARKET_CAP].values
x = ml_data.drop(MARKET_CAP, axis=1).values
x_train, x_test, y_train, y_test, = train_test_split(x, y)
# Step 1: Identify the columns containing Timestamp data
timestamp_columns = [col for col in ml_data.columns if ml_data[col].dtype == 'datetime64[ns]']

# Step 2: Convert the Timestamp values to Unix timestamp (seconds since 1970-01-01 00:00:00 UTC)
for col in timestamp_columns:
    ml_data[col] = ml_data[col].apply(lambda x: x.timestamp())

# Steps 3 and 4: Replace the original Timestamp columns with the converted numerical values and continue with the model training
x = ml_data.drop(MARKET_CAP, axis=1).values
x_train, x_test, y_train, y_test, = train_test_split(x, y)
model.fit(x_train, y_train)

predicted = model.predict(x_test)
print(r2_score(y_test, predicted))
```

```python
from trade_stat_logger.logger import SimpleLogger
from sklearn.preprocessing import LabelEncoder
logger = SimpleLogger()

latest_df = data5[[TICKER, DATE, MARKET_CAP]]
latest_df = latest_df[(latest_df[DATE].dt.year == 2023) & (latest_df[DATE].dt.month == 5)]
latest_df.set_index(TICKER, inplace=True)



backtest_data

def predict(data):
    return model.predict([data])[0]

for index in backtest_data.index:
    try:
        row = backtest_data.loc[index].copy().fillna(0)
        ticker = row[TICKER]
        mktcap = row[MARKET_CAP]
        mktcap_current = latest_df.loc[ticker][MARKET_CAP]
    except KeyError:
        break
    position = mktcap/1e6
    position_current = mktcap_current/1e6
    del row[TICKER], row[MARKET_CAP]
    # Step 1: Identify the columns containing Timestamp data
timestamp_columns = [col for col in backtest_data.columns if backtest_data[col].dtype == 'datetime64[ns]']

for col in timestamp_columns:
    backtest_data[col] = backtest_data[col].apply(lambda x: x.timestamp())

# Step 2: Convert the Timestamp values to Unix timestamp (seconds since 1970-01-01 00:00:00 UTC)
for row in timestamp_columns:
    backtest_data.loc[:, row] = backtest_data[row].apply(lambda x: x.timestamp())

    categorical_columns = [col for col in backtest_data.columns if backtest_data[col].dtype == 'object']

for column in categorical_columns:
    le = LabelEncoder()
    backtest_data.loc[:, column] = le.fit_transform(backtest_data[column])

# Steps 3 and 4: Replace the original Timestamp columns with the converted numerical values and continue with the model training
    predicted = predict(backtest_data.loc[index])
    if position > 0.0 and position_current > 0:
        if predicted * 0.9 < mktcap:
            logger.log(security=ticker, shares=1, share_price=mktcap)
            logger.log(security=ticker, shares=1, share_price=mktcap_current)
        elif predicted * 1.1 > mktcap:
            logger.log(security=ticker, shares=1, share_price=mktcap)
            logger.log(security=ticker, shares=1, share_price=mktcap_current)
logger.graph_statistics()
```
